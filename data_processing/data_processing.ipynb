```python
import matplotlib as plt
# Load the CSV data into a DataFrame


# Summary statistics
summary_stats = data_frame[['ER', 'CTR']].describe()
print(summary_stats)

# Distribution of ER and CTR
er_histogram = data_frame['ER'].hist()
er_histogram.set_title('Distribution of Engagement Rate (ER)')
er_histogram.set_xlabel('ER')
er_histogram.set_ylabel('Frequency')

ctr_histogram = data_frame['CTR'].hist()
ctr_histogram.set_title('Distribution of Click Through Rate (CTR)')
ctr_histogram.set_xlabel('CTR')
ctr_histogram.set_ylabel('Frequency')

# Correlation between ER and CTR
correlation = data_frame[['ER', 'CTR']].corr()
correlation
```

```python
import matplotlib.pyplot as plt
# Sort the DataFrame by ER and CTR in descending order
top_10_ctr = data_frame.sort_values('CTR', ascending=False).head(10)



# Plotting the top 10 ads with highest CTR
plt.figure(figsize=(10, 6))
plt.bar(top_10_ctr['game_id'], top_10_ctr['CTR'])
plt.xlabel('Game ID')
plt.ylabel('Click Through Rate (CTR)')
plt.title('Top 10 Ads with Highest Click Through Rate (CTR)')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_er = data_frame.sort_values('ER', ascending=False).head(10)


plt.figure(figsize=(10, 6))
plt.bar(top_10_er['game_id'], top_10_er['ER'])
plt.xlabel('Game ID')
plt.ylabel('Engagement Rate (ER)')
plt.title('Top 10 Ads with Highest Engagement Rate (ER)')
plt.xticks(rotation=90)
plt.show()
```
